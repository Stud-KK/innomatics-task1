## 2. Pass / Fail Analyzer
**Objective:**  
Analyze student marks and count the number of passed and failed students.


```python
# Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

## 3. Simple Data Cleaner
**Objective:**  
Clean and standardize user names by removing extra spaces and converting them to lowercase.


```python
# Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())

print("Cleaned Names List:", cleaned_names)
```

## 4. Message Length Analyzer
**Objective:**  
Analyze the length of each message and flag messages longer than 10 characters.


```python
# Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("⚠️ Flag: Message longer than 10 characters")
```

## 5. Error Message Detector
**Objective:**  
Detect and count the number of error messages from system logs.


```python
# Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
